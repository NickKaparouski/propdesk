```python
######## DZIAŁA ###########

import requests
from bs4 import BeautifulSoup

def get_page_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Проверка на успешный HTTP-статус
        return response.content
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {url}")
        print(e)
        return None

def is_real_estate_offer(link):
    # Check if the link contains "/nieruchomosci" in its URL
    excluded_urls = [
        'https://gratka.pl/nieruchomosci/mieszkania/warszawa',
        'https://gratka.pl/nieruchomosci/lokale-uzytkowe/warszawa',
        'https://gratka.pl/nieruchomosci/domy/warszawa',
        'https://gratka.pl/nieruchomosci/dzialki-grunty/warszawa',
        'https://gratka.pl/nieruchomosci/pokoje/warszawa',
        'https://gratka.pl/nieruchomosci/garaze/warszawa',
        'https://gratka.pl/nieruchomosci/warszawa/bialoleka',
        'https://gratka.pl/nieruchomosci/warszawa/mokotow',
        'https://gratka.pl/nieruchomosci/warszawa/wola',
        'https://gratka.pl/nieruchomosci/warszawa/ursynow',
        'https://gratka.pl/nieruchomosci/warszawa/praga-poludnie',
        'https://gratka.pl/nieruchomosci/warszawa/wlochy',
        'https://gratka.pl/nieruchomosci/warszawa/ursus',
        'https://gratka.pl/nieruchomosci/warszawa/wawer',
        'https://gratka.pl/nieruchomosci/warszawa/wilanow',
        'https://gratka.pl/nieruchomosci/warszawa/srodmiescie',
        'https://gratka.pl/nieruchomosci/warszawa/praga-polnoc',
        'https://gratka.pl/nieruchomosci/warszawa/bemowo',
        'https://gratka.pl/nieruchomosci/warszawa/targowek',
        'https://gratka.pl/nieruchomosci/warszawa/bielany',
        'https://gratka.pl/nieruchomosci/warszawa/ochota',
        'https://gratka.pl/nieruchomosci/warszawa/goclaw',
        'https://gratka.pl/nieruchomosci/warszawa/szamoty',
        'https://gratka.pl/nieruchomosci/warszawa/zoliborz',
        'https://gratka.pl/nieruchomosci/warszawa/tarchomin',
        'https://gratka.pl/nieruchomosci/warszawa/kabaty',
        'https://gratka.pl/nieruchomosci/warszawa/rembertow',
        'https://gratka.pl/nieruchomosci/warszawa/wesola',
        'https://gratka.pl/nieruchomosci/warszawa/sadyba',
        'https://gratka.pl/nieruchomosci/warszawa/zawady',
        'https://gratka.pl/nieruchomosci/warszawa/powsin',
        'https://gratka.pl/nieruchomosci/warszawa/powisle',
        'https://gratka.pl/nieruchomosci/warszawa/brodno',
        'https://gratka.pl/nieruchomosci/warszawa/saska-kepa',
        'https://gratka.pl/nieruchomosci/warszawa/zacisze',
        'https://gratka.pl/nieruchomosci/warszawa/grochow',
        'https://gratka.pl/nieruchomosci/grodzisk-mazowiecki',
        'https://gratka.pl/nieruchomosci/pruszkow',
        'https://gratka.pl/nieruchomosci/radom',
        'https://gratka.pl/nieruchomosci/marki',
        'https://gratka.pl/nieruchomosci/milanowek',
        'https://gratka.pl/nieruchomosci/konstancin-jeziorna',
        'https://gratka.pl/nieruchomosci/piaseczno',
        'https://gratka.pl/nieruchomosci/zabki',
        'https://gratka.pl/nieruchomosci/jaktorow',
        'https://gratka.pl/nieruchomosci/jozefow',
        'https://gratka.pl/nieruchomosci/radzymin',
        'https://gratka.pl/nieruchomosci/otwock',
        'https://gratka.pl/nieruchomosci/brwinow',
        'https://gratka.pl/nieruchomosci/osowiec-9782',
        'https://gratka.pl/nieruchomosci/lomianki',
        'https://gratka.pl/nieruchomosci/zyrardow',
        'https://gratka.pl/nieruchomosci/nadarzyn',
        'https://gratka.pl/nieruchomosci/legionowo',
        'https://gratka.pl/nieruchomosci/sulejowek',
        'https://gratka.pl/nieruchomosci/jozefoslaw',
        'https://gratka.pl/nieruchomosci/kobylka',
        'https://gratka.pl/nieruchomosci/blonie',
        'https://gratka.pl/nieruchomosci/stara-wies-34092',
        'https://gratka.pl/nieruchomosci/piastow',
        'https://gratka.pl/nieruchomosci/rusiec-34091',
        'https://gratka.pl/nieruchomosci/minsk-mazowiecki',
        'https://gratka.pl/nieruchomosci/ozarow-mazowiecki',
        'https://gratka.pl/nieruchomosci/zabia-wola',
        'https://gratka.pl/nieruchomosci/serock',
        'https://gratka.pl/nieruchomosci/wolomin',
        'https://gratka.pl/blog/nieruchomosci/',
        'https://gratka.pl/oferta/klient-biznesowy/nieruchomosci/',
        'https://gratka.pl/blog/',
        'https://gratka.pl/mapa/nieruchomosci/warszawa',
        'https://gratka.pl/nieruchomosci/warszawa',
        'https://gratka.pl/nieruchomosci/warszawa?'
    ]

    excluded_words = ["dom", "inwestycje", "lokal", "rezerwa", "dzia", 'bli']

    return (
    "/nieruchomosci" in link
    and "/#contactWrapper" not in link
    and not any(url in link for url in excluded_urls)
    and not link == "https://gratka.pl/nieruchomosci"
    and not any(word in link for word in excluded_words)
)



def parse_page(url):
    content = get_page_content(url)
    scraped_links = []
    if content is not None:
        soup = BeautifulSoup(content, 'html.parser')
        links = soup.find_all('a')
        for link in links:
            url = link.get('href')
            if url and is_real_estate_offer(url):
                scraped_links.append(url)
    return scraped_links

def get_next_page_url(current_url, page):
    # This function should be implemented to find and return the URL of the next page.
    # The implementation depends on the website's specific pagination structure.
    # You can use BeautifulSoup to find the "Next" button or other pagination controls
    # and extract the URL of the next page.
    # Replace the placeholder implementation below with your actual logic.
    return "/nieruchomosci/warszawa?page=" + str(page) in current_url

if __name__ == "__main__":
    base_url = "https://gratka.pl/nieruchomosci/warszawa?page="
    page_number = 1
    all_scraped_links = []

    while True:
        url_to_crawl = base_url + str(page_number)
        print(f"Scraping page {page_number} - {url_to_crawl}")
        scraped_links = parse_page(url_to_crawl)
        all_scraped_links.extend(scraped_links)

        next_page_url = get_next_page_url(url_to_crawl, (page_number + 1))
        if next_page_url is None:
            # If there are no more pages or unable to find the "Next" button, stop crawling
            print("No more pages or unable to find the 'Next' button.")
            break

        # Update the page_number to crawl the next page
        page_number += 1
    print('All scraped links:')
    for link in all_scraped_links:
        print(link)
```
